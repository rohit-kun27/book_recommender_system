## Popularity Based Recommender System

```python
books
```

```python
ratings_with_name = ratings.merge(books, on='isbn')
```

```python
num_rating_df = ratings_with_name.groupby('book_title')['rating'].count().reset_index()
num_rating_df.rename(columns={'rating': 'num_rating'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('book_title')['rating'].mean().round(1).reset_index()
avg_rating_df.rename(columns={'rating': 'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on= 'book_title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=150].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on ='book_title').drop_duplicates('book_title')[['book_title','book_author','num_rating','avg_rating']]
```

```python
popular_df.shape
```

## Collabarative Filtering Model 

```python
x = ratings_with_name.groupby('user_id').count()['rating'] > 30
exp_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['user_id'].isin(exp_users)]
```

```python
y = filtered_rating.groupby('book_title').count()['rating'] >= 5
famous_books = y[y].index
```

```python
famous_books
```

```python
final_rating = filtered_rating[filtered_rating['book_title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='book_title',columns='user_id',values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
import numpy as np
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['book_title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('book_title')['book_title'].values))
        item.extend(list(temp_df.drop_duplicates('book_author')['book_author'].values))

        data.append(item)

    return data
    
    
```

```python
recommend('1984')
```

```python
recommend('The Da Vinci Code')
```

## Demographic-Based Recommender System

```python
new_df = ratings.merge(users, on='user_id').merge(books, on='isbn')
```

```python
new_df
```

```python
new_df['age_group'] = pd.cut(new_df['Age'], bins=[10, 18, 25, 35, 50, 70, 100],
                         labels=['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior', 'Elder'])
```

```python
new_df
```

```python
new_df['age_group'].unique()
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(subset=['age_group'], inplace=True) 
```

```python
new_df.isnull().sum()
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
def recommend_by_age(age, top_n):   #A Function to recommend top N books for a given age
    if age < 10 or age > 100:
        return "Age out of supported range (10–100)."
    
    age_group = new_df[new_df['Age'] == age]['age_group'].values
    if len(age_group) == 0:
        return "No users found in this age group."
    
    group = age_group[0]
    group_df = new_df[new_df['age_group'] == group]

    book_stats = group_df.groupby(['isbn', 'book_title']).agg(  # Taking average ratings & count
        avgg_rating=('rating', 'mean'),
        rating_count=('rating', 'count')
    ).reset_index()

    popular_books = book_stats[book_stats['rating_count'] >= 10] # only considering books with 10+ ratings in group
    top_books = popular_books.sort_values('avgg_rating', ascending=False).head(top_n)
    
    return top_books[['book_title', 'avgg_rating', 'rating_count']]
```

```python
recommend_by_age(14,5)
```
